```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Bovina = [22.1, 22.1, 23.1, 23.9, 24.6, 21.7, 23.5, 22.0, 22.5, 23.6, 21.7]
Tiempo = [2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011]

Porcina = [17.9, 19.4, 19.1, 18.3, 19.3, 22.5, 23.5, 25.0, 24.0, 24.4, 25.6]
Tiempo = [2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011]
```

```python
plt.figure(figsize=[10,8])
plt.plot(Tiempo, Bovina, marker='o', linestyle='--', color='b', label='Carne bovina')
plt.plot(Tiempo, Porcina, marker='d', linestyle='-', color='r', label='Carne porcina')
plt.ylabel("Carne (kg/hab)")
plt.xlabel("Aos")
plt.title("Consumo de carne anual")
plt.legend(loc="lower right")
plt.show
```

```python
plt.figure(figsize=[8,8])
Autos = ["Cia", "Toyota", "Nissan", "Suzuki", "Audi"]
Venta = [10.5, 13.5, 14.2, 16.1, 9.8]
Destacar = (0.1,0,0,0,0)
plt.style.use("ggplot")
plt.title("Ventas de autos")
plt.pie(x=Venta, explode=Destacar, labels=Autos, autopct="%.2f%%", shadow=True, startangle=20)
plt.axis = ("equal")
plt.legend(loc = "upper left")
plt.show()
```

```python
plt.figure(figsize=[8,8])

Toyota = (10, 19, 15, 14, 9)
Audi = (15, 25, 35, 15, 10)

Espacio = [0,1,2,3,4]
Grosor = 0.5

plt.title("Venta de autos")
plt.xticks(Espacio,("Dia 1", "Dia 2", "Dia 3", "Dia 4", "Dia 5"))

Bar1 = plt.bar(Espacio, Toyota, Grosor)
Bar2 = plt.bar(Espacio, Audi, Grosor, bottom = Toyota)

plt.ylabel("Unidades vendidas")
plt.yticks([0,5,10,15,20,25,30,35,40])
plt.legend((Bar1[0], Bar2[0]), ("Toyota","Audi"))
plt.show()
```

```python
plt.figure(figsize=[8,8])

Audi = (15, 25, 35, 15, 10)
AudiReal = (12, 28, 29, 10, 9)
Espacio = [0,1,2,3,4]
Grosor = 0.5

plt.title("Venta de autos")
plt.xticks(Espacio,("Dia 1", "Dia 2", "Dia 3", "Dia 4", "Dia 5"))

Bar1 = plt.plot(Espacio, Audi, Grosor, linestyle="--", marker='d', color="b")
Bar2 = plt.bar(Espacio, AudiReal, Grosor)

plt.ylabel("Unidades vendidas")
plt.yticks([0,5,10,15,20,25,30,35,40])
plt.show()
```

```python
n = 100
x = np.linspace(0,10,n)
y = x + np.random.rand(n)

fig = plt.figure()
ax = fig.add_subplot()
im = ax.scatter(x, y, s=100, c=y, alpha=0.5, edgecolors="black",cmap="viridis")
cbar = fig.colorbar(im)
cbar.set_label("Temperatura (C)")
ax.set_xlabel("Tiempo (min)", fontsize=14)
ax.set_ylabel("Bacterias (miles)", fontsize=14)
ax.grid(alpha=0.25)
plt.show
```
